```python
# importare le librerie necessarie
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# carica il tuo dataset
df = pd.read_csv(r"C:\Users\manet\OneDrive\Desktop\Cancer Breast data\data.csv")

#elimino attributo non numerico e id (inutile per correlazione)
df = df.drop('diagnosis', axis=1)
df = df.drop('id', axis=1)

# calcola la matrice di correlazione
corr = np.corrcoef(df.values.T)

# utilizzare la funzione "heatmap" di seaborn per creare un heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(corr, xticklabels=df.columns, yticklabels=df.columns, annot=True)
plt.savefig('heatmap1.png')
```

```python
drop_list1 = ['perimeter_mean','radius_mean','radius_se','perimeter_se','radius_worst','perimeter_worst']
df = df.drop(drop_list1, axis=1)

# calcola la matrice di correlazione
corr = np.corrcoef(df.values.T)

# utilizzare la funzione "heatmap" di seaborn per creare un heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(corr, xticklabels=df.columns, yticklabels=df.columns, annot=True)
plt.savefig('heatmap2.png')
```

```python
drop_list2 = ['concavity_mean','compactness_mean','concavity_se','compactness_se','concavity_worst','compactness_worst']
df = df.drop(drop_list2, axis=1)

# calcola la matrice di correlazione
corr = np.corrcoef(df.values.T)

# utilizzare la funzione "heatmap" di seaborn per creare un heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(corr, xticklabels=df.columns, yticklabels=df.columns, annot=True)
plt.savefig('heatmap3.png')
```
